```python
def random_date_generator(start_date, range_in_days):
    days_to_add = np.arange(0, range_in_days)
    random_date = np.datetime64(start_date) + np.random.choice(days_to_add)
    return pd.to_datetime(random_date)
```

```python
data = pd.DataFrame()

for i in list(zip(users, months)):
    
    start = random_date_generator('2017-01-01', 365)
    time_range = [start + pd.DateOffset(months=month) for month in range(i[1])]

    df = pd.DataFrame({'user_id': [i[0] for _ in range(i[1])],
                       'action': ['subscribed'] + ['renewed' for _ in range(i[1]-2)] + ['cancelled'],
                       'payment': [199 for _ in range(i[1]-1)] + [0],
                       'date': time_range
                       })
    data = data.append(df)
```

```python
data = pd.DataFrame({'user_id': np.arange(1, 10001),
                     'months_of_subscription': np.random.choice(month_number, 10000)})
```

```python
actions = {}
for index, row in data.iterrows():
    n = row['months_of_subscription']
    m = row['user_id']
    actions[m] = []
    actions[m].append('subscribed')
    for i in range(n-2):
        actions[m].append('renewed')
    actions[m].append('cancelled')
        
```
